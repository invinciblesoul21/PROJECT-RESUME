```python
import pandas as pd

# Load the dataset
file_path = "Global_Cybersecurity_Threats_2015-2024.csv"

try:
    df = pd.read_csv(file_path)
    print("✅ File loaded successfully!")
    display(df.head())  # show first 5 rows
except FileNotFoundError:
    print("❌ Error: File not found.")
except pd.errors.EmptyDataError:
    print("❌ Error: File is empty.")
except pd.errors.ParserError:
    print("❌ Error: File is corrupted or not a valid CSV.")
```

```python
df.rename(columns={'Incident Resolution Time (in Hours)':'Incident Resolution Status (in Hours)'},inplace=True)
df
```

```python
df
```

```python
#Which Attack Type causes the highest average financial loss?

a=df.groupby('Attack Type')['Financial Loss (in Million $)'].mean()
a
```

```python
a=df.groupby('Attack Type')['Financial Loss (in Million $)'].mean().idxmax()
a
```

```python
a=df.groupby('Attack Type')['Financial Loss (in Million $)'].mean().idxmin()
a
```

```python
#Is there a correlation between the number of affected users and financial loss?
import seaborn as sns
import matplotlib.pyplot as plt

sns.scatterplot(data=df, x='Financial Loss (in Million $)', y='Number of Affected Users',color='r')
plt.title('Scatter Plot')
plt.show()
```

```python
b=df['Financial Loss (in Million $)'].corr(df['Number of Affected Users'])
b
```

```python
#Which countries have faced the most incidents overall?

sns.barplot(data=df,x='Country',y='Number of Affected Users')
plt.xticks(rotation=60)
plt.xlabel('c')
plt.ylabel('NAU')
plt.show()
```

```python
#What are the top 3 most common defense mechanisms, and how effective are they in reducing resolution time?

d=df.groupby('Defense Mechanism Used')['Incident Resolution Status (in Hours)'].mean()
d #Most effective is Firewall
```

```python
d=df.groupby('Defense Mechanism Used')['Incident Resolution Status (in Hours)'].mean().idxmin()
d #Most effective is Firewall
```

```python
d=df.groupby('Defense Mechanism Used')['Incident Resolution Status (in Hours)'].mean().idxmax()
d #Most effective is Firewall
```

```python
#Has the nature of attacks changed over the years (2015–2024)?
import seaborn as sns
import matplotlib.pyplot as plt
sns.lineplot(data=df,x='Year',y='Attack Type',color='y')
plt.xticks(rotation=45)
plt.show()
```

```python
attack_counts = df['Attack Type'].value_counts()

plt.figure(figsize=(6,6))
plt.pie(attack_counts, labels=attack_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Attack Types')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x='Attack Type', y='Financial Loss (in Million $)', data=df)
plt.xticks(rotation=45)
plt.title('Financial Loss Distribution by Attack Type')
plt.show()
```
